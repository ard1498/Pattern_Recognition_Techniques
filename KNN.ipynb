# KNN Implementation

```python
import sklearn.datasets as dt
import numpy as np
from sklearn.metrics import accuracy_score
from collections import Counter
```

```python
iris = dt.load_iris()
X = iris.data
Y = iris.target
```

```python
from sklearn import model_selection as ms
```

```python
xtrain, xtest, ytrain, ytest = ms.train_test_split(X, Y)
xtrain.shape, ytrain.shape, xtest.shape, ytest.shape
```

```python
class KNN_Neighbor:
    
    def __init__(self, xtrain, ytrain, k = 0):
        self.xtrain = xtrain
        self.ytrain = ytrain
        self.k = k
        return
    
    def predict_one(self, xtest):
        distances = []
        for i in range(len(self.xtrain)):
            distance = ((self.xtrain[i, :] - xtest)**2).sum()
            distances.append([distance, i])
        distances = sorted(distances)
        k_neighbours = []
        print("For xtest : ")
        pprint(xtest)
        print(" ---> ")
        for i in range(self.k):
            k_neighbours.append(self.ytrain[distances[i][1]])
            print(f"{[distances[i][0]]}, {[distances[i][1]]}")
            
        print("---------------------------------------------")
        return Counter(k_neighbours).most_common(1)[0][0]
    
    def predict(self, xtest_data):
        ypred = []
        for xtesti in xtest_data:
            ypred.append(self.predict_one(xtesti))
        return ypred
```

```python
k = int(input('Enter the no of neighbors to choose :'))
clf = KNN_Neighbor(xtrain, ytrain, k)
Ypred = clf.predict(xtest)
accuracy_score(ytest, Ypred)
```
